```python
# 1 – Verificar se a pessoa pode votar
# Recebe a idade do usuário e verifica se ela pode votar

idade = int(input("Qual sua idade? "))

if idade >= 16:
    print("Você pode votar!")
else:
    print("Você não pode votar ainda.")

```

```python
# 2 – Verificar se o aluno está aprovado ou reprovado
# Recebe as duas notas e calcula a média para verificar aprovação

nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
media = (nota1 + nota2) / 2

if media >= 7:
    print("Aluno aprovado!")
else:
    print("Aluno reprovado!")
```

```python
# 3 – Classificar a sensação térmica
# Recebe a temperatura e classifica como fria, agradável, quente ou muito quente

temperatura = float(input("Digite a temperatura (em °C): "))

if temperatura < 15:
    print("Está frio.")
elif 15 <= temperatura < 25:
    print("Está agradável.")
elif 25 <= temperatura < 35:
    print("Está quente.")
else:
    print("Está muito quente!")
```

```python
# 4 – Verificar se o número é par ou ímpar
# Recebe um número inteiro e verifica se é par ou ímpar

numero = int(input("Digite um número inteiro: "))

if numero % 2 == 0:
    print("O número é par.")
else:
    print("O número é ímpar.")
```

```python
# 5 – Reajuste de salários
# Recebe o salário e aplica o aumento conforme as faixas salariais

salario = float(input("Digite o seu salário atual: "))

if salario <= 1500:
    aumento = salario * 0.10
elif salario <= 3000:
    aumento = salario * 0.07
else:
    aumento = salario * 0.05

novo_salario = salario + aumento
print(f"Seu novo salário é: R${novo_salario:.2f}")
```

```python
# 6 – Identificar o maior número
# Recebe três números e exibe o maior

numero1 = int(input("Digite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))
numero3 = int(input("Digite o terceiro número: "))

maior = max(numero1, numero2, numero3)
print(f"O maior número é: {maior}")
```

```python
# 7 – Verificar se o ano é bissexto
# Recebe um ano e verifica se ele é bissexto

ano = int(input("Digite um ano para verificar se é bissexto: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

```python
# 8 – Determinar o tipo de ingresso
# Recebe a idade e define o tipo de ingresso: infantil ou adulto

idade = int(input("Digite a sua idade: "))

if idade < 18:
    print("Ingresso infantil.")
else:
    print("Ingresso adulto.")
```

```python
# 9 – Identificar o produto pelo código
# Recebe um código e exibe o produto correspondente

codigo = int(input("Digite o código do produto (1, 2 ou 3): "))

if codigo == 1:
    print("Produto: Maçã")
elif codigo == 2:
    print("Produto: Banana")
elif codigo == 3:
    print("Produto: Laranja")
else:
    print("Código inválido.")
```

```python
# 10 – Exibir saudação conforme o período do dia
# Recebe o horário e exibe uma saudação adequada

hora = int(input("Digite o horário (0 a 23): "))

if 6 <= hora < 12:
    print("Bom dia!")
elif 12 <= hora < 18:
    print("Boa tarde!")
elif 18 <= hora < 24:
    print("Boa noite!")
else:
    print("Hora inválida!")
```

```python
# 11 – Soma de números até 0 ser digitado
# Continua somando números até que o usuário digite 0

soma = 0
while True:
    numero = int(input("Digite um número (0 para parar): "))
    if numero == 0:
        break
    soma += numero

print(f"A soma total dos números é: {soma}")
```

```python
# 12 – Sistema de login
# Continua pedindo a senha até que o usuário digite "python123"

senha = ""
while senha != "python123":
    senha = input("Digite a senha: ")

print("Bem-vindo ao sistema!")
```

```python
# 13 – Soma de números sequenciais até ultrapassar 100
# Soma números começando de 1 até que a soma seja maior que 100

soma = 0
numero = 1

while soma <= 100:
    soma += numero
    print(f"Adicionando {numero}. Soma atual: {soma}")
    numero += 1
```

```python
# 14 – Média das idades digitadas
# Continua lendo idades até que um número negativo seja digitado e calcula a média

idades = []
while True:
    idade = int(input("Digite uma idade (negativo para parar): "))
    if idade < 0:
        break
    idades.append(idade)

media = sum(idades) / len(idades) if idades else 0
print(f"A média das idades é: {media:.2f}")
```

```python
# 15 – Contagem regressiva
# Realiza uma contagem regressiva a partir de um número informado

numero = int(input("Digite o número inicial para a contagem regressiva: "))

while numero >= 0:
    print(numero)
    numero -= 1

```

```python
# 19 – Contar idades maiores que 18
# Conta quantas idades digitadas são maiores que 18

contagem = 0
for i in range(5):
    idade = int(input(f"Digite a {i+1}ª idade: "))
    if idade > 18:
        contagem += 1

print(f"Quantidade de idades maiores que 18: {contagem}")
```

```python
# 20 – Encontrar o maior número
# Recebe 7 números e determina o maior deles

maior = None
for i in range(7):
    numero = int(input(f"Digite o {i+1}º número: "))
    if maior is None or numero > maior:
        maior = numero

print(f"O maior número é: {maior}")
8
```

```python

# 21 – Encontrar a tarefa com o nome mais longo
# Recebe 4 nomes de tarefas e mostra o mais longo

tarefas = []
for i in range(4):
    tarefa = input(f"Digite o nome da {i+1}ª tarefa: ")
    tarefas.append(tarefa)

tarefa_mais_longa = max(tarefas, key=len)
print(f"A tarefa com o nome mais longo é: {tarefa_mais_longa}")
```

```python
# 22 – Somar números pares entre 2 e 100
# Soma todos os números pares de 2 até 100

soma = 0
for numero in range(2, 101, 2):
    soma += numero

print(f"A soma dos números pares entre 2 e 100 é: {soma}")
```

```python
# 23 – Contar positivos, negativos e zeros
# Recebe 10 números e conta quantos são positivos, negativos ou iguais a zero

positivos = negativos = zeros = 0
for i in range(10):
    numero = int(input(f"Digite o {i+1}º número: "))
    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Positivos: {positivos}, Negativos: {negativos}, Zeros: {zeros}")

```

```python
# 24 – Filtrar nomes que começam com "A" ou "C"
# Lê 6 nomes e exibe os que começam com "A" ou "C"

for i in range(6):
    nome = input(f"Digite o {i+1}º nome: ")
    if nome[0].upper() in ['A', 'C']:
        print(nome)
```

```python
# 25 – Caixa registradora digital
# Continua lendo números até que 999 seja digitado

while True:
    valor = float(input("Digite um valor (999 para encerrar): "))
    if valor == 999:
        break
    print(f"Valor inserido: R${valor:.2f}")
```

```python
# 26 – Listar números ignorando múltiplos de 4
# Exibe números de 1 a 20, mas ignora múltiplos de 4

for numero in range(1, 21):
    if numero % 4 == 0:
        continue
    print(numero)
```

```python
# 27 – Sistema de login com limite de tentativas
# O usuário tem no máximo 3 tentativas para acertar a senha

tentativas = 0
senha = ""
while senha != "python123" and tentativas < 3:
    senha = input("Digite a senha: ")
    tentativas += 1

if senha == "python123":
    print("Bem-vindo ao sistema!")
else:
    print("Número máximo de tentativas excedido.")
```

```python
# 28 – Somar apenas números positivos
# Soma os números positivos digitados, ignorando negativos

soma = 0
for i in range(10):
    numero = float(input(f"Digite o {i+1}º número: "))
    if numero < 0:
        continue
    soma += numero

print(f"A soma dos números positivos é: {soma}")
1
```

```python
# 29 – Acumulador até a soma ultrapassar 200
# Soma números até que o total ultrapasse 200

soma = 0
while soma <= 200:
    numero = int(input("Digite um número: "))
    soma += numero

print(f"A soma total é: {soma}")
```

```python
# 30 – Contar palavras válidas
# Conta quantas palavras foram digitadas até o usuário digitar "fim"

contador = 0
while True:
    palavra = input("Digite uma palavra (ou 'fim' para terminar): ")
    if palavra.lower() == "fim":
        break
    contador += 1

print(f"Você digitou {contador} palavras.")
```
