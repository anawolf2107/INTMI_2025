```python
#-Verificar se o número é maior que 10
numero = float(input("Digite um número: "))

if numero > 10:
    print("O número é maior que 10.")
else:
    print("O número não é maior que 10.")
```

```python
#-Verificar se o número é par
numero = int(input("Digite um número inteiro: "))

if numero % 2 == 0:
    print("O número é par.")
else:
    print("O número é ímpar.")
```

```python
#-Verificar se o nome contém a letra 'a'
nome = input("Digite um nome: ")

if "a" in nome.lower():
    print("O nome contém a letra 'a'.")
else:
    print("O nome não contém a letra 'a'.")
```

```python
#-Verificar se o número é múltiplo de 5
numero = float(input("Digite um número: "))

if numero % 5 == 0:
    print("O número é múltiplo de 5.")
else:
    print("O número não é múltiplo de 5.")
```

```python
#-Verificar se está quente ou frio
temperatura = float(input("Digite a temperatura atual (°C): "))

if temperatura > 30:
    print("Está quente!")
else:
    print("Está frio!")
```

```python
#-Verificar se o usuário quer continuar ou sair
opcao = input('Digite "S" para continuar ou "N" para sair: ').strip().upper()

if opcao == "S":
    print("Você escolheu continuar!")
elif opcao == "N":
    print("Você escolheu sair. Até mais!")
else:
    print("Opção inválida. Digite apenas 'S' ou 'N'.")
```

```python
#-Classificar número como negativo, zero ou positivo
numero = float(input("Digite um número: "))

if numero < 0:
    print("O número é negativo.")
elif numero == 0:
    print("O número é zero.")
else:
    print("O número é positivo.")
```

```python
#-Número de 1 a 7 → dia da semana
numero = int(input("Digite um número de 1 a 7: "))

dias = {
    1: "Domingo",
    2: "Segunda-feira",
    3: "Terça-feira",
    4: "Quarta-feira",
    5: "Quinta-feira",
    6: "Sexta-feira",
    7: "Sábado"
}

if 1 <= numero <= 7:
    print(f"O dia da semana é: {dias[numero]}")
else:
    print("Número inválido! Digite um número entre 1 e 7.")
```

```python
#- Verificar se o ano é bissexto
ano = int(input("Digite um ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é bissexto.")
else:
    print(f"O ano {ano} não é bissexto.")
```

```python
#-Calcular salário com hora extra
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_hora) + (horas_extras * valor_hora * 1.5)

print(f"Seu salário é R$ {salario:.2f}")
```
